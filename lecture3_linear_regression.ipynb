```python
X = list(range(10))
Y = list(range(1,20,2))
```

```python
Y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
plt.plot(X,Y)
```

```python
class H():
  def __init__(self, w):
    self.w = w
  
  def forward(self, x):
    return self.w * x

def cost(h, X, Y):
  error = 0
  for i in range(len(X)):
    error += (h.forward(X[i]) - Y[i])**2
  error = error / len(X)
  return error
```

```python
h = H(2.1)
cost(h, X, Y)
```

```python
list_w = []
list_c = []
for i in range(-100, 100):
  w = i * 0.1
  h = H(w)
  c = cost(h, X, Y)
  list_w.append(w)
  list_c.append(c)
```

```python
print(list_w)
print(list_c)
```

```python
plt.figure(figsize = (10,10))
plt.xlabel("weight")
plt.ylabel("cost")
plt.plot(list_w, list_c)
plt.scatter(list_w, list_c, s = 10)
```

```python
def cal_grad1(w):
  eps = 0.001

  h = H(w)
  cost1 = cost(h, X, Y)
  h = H(w + eps)
  cost2 =  cost(h, X, Y)

  dcost = cost2 - cost1 
  dw = eps

  grad = dcost/dw
  return grad, (cost1 + cost2)*0.5

def cal_grad2(w):
  h = H(w)
  grad = 0
  for i in range(len(X)):
    grad += 2*(h.forward(X[i]) - Y[i]) * X[i]
  grad = grad / len(X)
  c = cost(h, X, Y)
  return grad, c

w1 = 4
w2 = 4
lr = 0.01

list_w1, list_w2 = [], []
list_c1, list_c2 = [], []

for i in range(10):
  grad1, mean_cost1 = cal_grad1(w1)
  grad2, mean_cost2 = cal_grad2(w2)

  list_w1.append(w1)
  list_w2.append(w2)
  list_c1.append(mean_cost1)
  list_c2.append(mean_cost2)

  w1 = w1 - lr * grad1
  w2 = w2 - lr * grad2

plt.plot(list_w1, list_c1) 
```
